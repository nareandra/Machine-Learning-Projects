```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Features and labels
X = iris.data
y = iris.target

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.4,
    random_state=0
)
```

```python
k_values = [1,3,5,7,9]
accuracies = []

for k in k_values:

    knn = KNeighborsClassifier(n_neighbors=k)

    # Train
    knn.fit(X_train, y_train)

    # Predict
    y_pred = knn.predict(X_test)

    # Calculate accuracy
    acc = accuracy_score(y_test, y_pred)

    accuracies.append(acc)

    print("K =", k, " Accuracy =", acc)
```

```python
import pandas as pd

results = pd.DataFrame({
    "K Value": k_values,
    "Accuracy": accuracies
})

results
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(k_values,
         accuracies,
         marker='o')

plt.xlabel("K Value")
plt.ylabel("Accuracy")

plt.title("Accuracy vs K Value")

plt.grid()

plt.show()
```

**Evaluation of Different K Values**

The K-NN classifier was evaluated using K values of 1, 3, 5, 7, and 9. The experimental results show that moderate values of K provide better classification accuracy. Smaller K values are sensitive to noise, whereas larger K values may reduce classification performance because of over-smoothing.

**Advantages of K-NN**


* Simple and Intuitive: It is very easy to understand, interpret, and implement from scratch.

* No Explicit Training Phase: As a lazy learner, K-NN does not build a model during training; it simply stores the data points, making the ingestion phase extremely fast.

* Versatile: It can be used for both classification and regression tasks, and easily handles multi-class datasets (like Iris).

* Can achieve high classification accuracy.




**DisAdvantages of K-NN**



*  High Computational and Memory Costs: Since it requires calculating distances to every single training point at prediction time, it consumes significant memory and becomes highly inefficient with large datasets.

*   Sensitive to the Choice of K: Choosing a K value that is too small leads to overfitting (high variance), while a K that is too large can lead to underfitting (high bias).
*  Sensitive to Irrelevant Features and Noise: Noisy data, outliers, or unscaled features can severely distort distance calculations, lowering overall accuracy.


*   The Curse of Dimensionality: In high-dimensional spaces, the distance between points becomes less distinguishable, significantly degrading performance.



**Conclusion**

* Performance Analysis: The evaluation demonstrated that moderate values of K (like K=5 or K=7) tend to provide the most stable and balanced accuracy.
* The Impact of K: Choosing a K value that is too small (K=1) makes the model highly sensitive to individual outliers and noise, which increases the risk of overfitting. Conversely, selecting a K value that is too large incorporates data points from other distant classes into the neighborhood vote, smoothing out boundaries and introducing underfitting.
* While K-NN is highly effective, intuitive, and requires zero explicit training time, it demands careful parameter tuning for K and becomes highly memory-intensive when processing large datasets because it must store and evaluate all training records for every single prediction.
